**Q3**: Make three dataframes that are sorted by season, winning team, and winning score respectively. Then, Using iloc, select the rows from index 100 to 200 and the columns for season, winning team, and winning score, respectively. 

To create the three dataframes sorted by season, winning team, and winning score, I start by using iloc. When using iloc, an extra value must be included; therefore, I use 100:201. This is applied to all three dataframes.

```python
# Write your code here

# Sorted by season
sorted_season = df.sort_values('Season')
dataframe_season = sorted_season.iloc[100:201]
dataframe_season
```

```python
# Sorted by winning team
sorted_wteam = df.sort_values('Wteam')
dataframe_wteam = sorted_wteam.iloc[100:201]
dataframe_wteam
```

```python
# Sorted by winning score
sorted_wscore = df.sort_values('Wscore')
dataframe_wscore = sorted_wscore.iloc[100:201]
dataframe_wscore
```

**Q4**: From these three subsets you obtained above, find the season and winning team for the game with the highest winning score.

To find the season and winning team with the highest winning score, I combine the three subsets created in Q3 and use the "idxmax()" function. Then, I extract the season, team, and score.

```python
# Combine the three subset from Q3, with the function "concat()", which is used to combine multiple DataFrames along a specified axis.
combined_subset = pd.concat([dataframe_season, dataframe_wteam, dataframe_wscore])

# Find the game with the highest score, with "idxmax()"
highest_score_row = combined_subset.loc[combined_subset['Wscore'].idxmax()]

season = highest_score_row['Season']
winning_team = highest_score_row['Wteam']
winning_score = highest_score_row['Wscore']

print(f"The season is {season} and the winning team is {winning_team} with the winning score of {winning_score}.")
```

**Q5**: Create a new column in the DataFrame called 'ScoreDifference' which is the absolute difference between the winning score and the losing score. Filter the DataFrame to only include games where the 'ScoreDifference' is greater than the average 'ScoreDifference' for all games.

The goal of this task is to create a new column in the dataframe called 'ScoreDifference', which represents the absolute difference between the winning team's score and the losing team's score.

First, the new column is created by taking the absolute value of winning score minus losing score.
Then, the average of 'ScoreDifference' is calculated using the "mean()" function.
Finally, a filter is applied to keep only the games where 'ScoreDifference' is greater than the average 'ScoreDifference'.

```python
# Write your code here

# Create the new column 'ScoreDifference'
df['ScoreDifference'] = abs(df['Wscore'] - df['Lscore'])

# Calculate the average of 'ScoreDifference'
average_score_difference = df['ScoreDifference'].mean()

# Make the filter in the Dataframe so it only includes games where 'ScoreDifference' is bigger than average
filter_df = df[df['ScoreDifference'] > average_score_difference]

filter_df
```

**Q6**: From this filtered DataFrame, find the season and teams involved in the game with the highest 'ScoreDifference'.

To find the season and the teams involved in the game with the largest 'scoredifference', I follow these steps:

Identify the row with the highest 'scoredifference' using the "idxmax()" function.
Extract the corresponding season, winning team, losing team, and scoredifference from that row.

```python
# Write your code here
# Find the row with the highest 'scoredifference' using the idxmax() function.
highest_diff_row = filter_df.loc[filter_df['ScoreDifference'].idxmax()]

season = highest_diff_row['Season']
winning_team = highest_diff_row['Wteam']
losing_team = highest_diff_row['Lteam']

print(f"The season is {season}, with the winning team {winning_team}, and the losing team {losing_team}.")
```

**Q14**: Calculate the total number of games played by each team, whether they won or lost. This will require iterating over the rows of the DataFrame and updating a dictionary that keeps track of the number of games for each team.

I made a code that calculates the total number of games played by each team, whether they won or lost.

Initialization: An empty dictionary (team_game_counts) is created to keep track of the number of games for each team.

Iterating through the DataFrame: The "iterrows()" function is used to loop through all rows in the DataFrame.

Updating game counts:

If a team is already in the dictionary, its count is incremented by 1.

If not, the team is added with an initial count of 1.

Result: The dictionary team_game_counts displays the total number of games played by each team.

```python
# Write your code here
# Initialize an empty dictionary to keep track of game counts
games_played = {}

# Iterate over the DataFrame rows
for index, row in df.iterrows():
    # Update counts for the winning team
    if row['Wteam'] in games_played:
        games_played[row['Wteam']] += 1
    else:
        games_played[row['Wteam']] = 1
    
    # Update counts for the losing team
    if row['Lteam'] in games_played:
        games_played[row['Lteam']] += 1
    else:
        games_played[row['Lteam']] = 1

games_played
```

**Q15**: For each season, find the game with the highest score difference (winning score - losing score). This will require iterating over the rows of the DataFrame, keeping track of the highest score difference for each season, and updating it if a game with a higher score difference is found.

This code finds the game with the highest score difference (winning score - losing score) for each season.

Initialize a dictionary: season_highest_difference is created to store the highest score difference for each season.

Iterate through the DataFrame: The code loops through each row using "iterrows()".

Calculate the score difference: The difference between the winning and losing score is computed.

Update the dictionary:

If the season is not in the dictionary, the current game is stored as the highest difference.

If the season already exists, the game is updated only if the current game has a higher score difference.

Convert to DataFrame: The dictionary is transformed into a DataFrame

Sorting: The final DataFrame is sorted by season for clarity.

```python
# Initialize a dictionary to track the highest score difference for each season
season_highest_difference = {}

# Iterate through each row of the DataFrame
for index, row in df.iterrows():
    # Calculate the score difference
    score_difference = row['Wscore'] - row['Lscore']
    
    # Get the current season
    season = row['Season']
    
    # Check if this season is already in the dictionary
    if season not in season_highest_difference:
        # If not, initialize with the current game
        season_highest_difference[season] = {
            'ScoreDifference': score_difference,
            'WinningTeam': row['Wteam'],
            'LosingTeam': row['Lteam'],
            'WinningScore': row['Wscore'],
            'LosingScore': row['Lscore']
        }
    else:
        # If it exists, check if the current game has a higher score difference
        if score_difference > season_highest_difference[season]['ScoreDifference']:
            season_highest_difference[season] = {
                'ScoreDifference': score_difference,
                'WinningTeam': row['Wteam'],
                'LosingTeam': row['Lteam'],
                'WinningScore': row['Wscore'],
                'LosingScore': row['Lscore']
            }

# Convert the dictionary to a DataFrame for better presentation
highest_diff_per_season = pd.DataFrame.from_dict(season_highest_difference, orient='index').reset_index()
highest_diff_per_season.columns = ['Season', 'ScoreDifference', 'WinningTeam', 'LosingTeam', 'WinningScore', 'LosingScore']

# Sort by season for better readability
highest_diff_per_season = highest_diff_per_season.sort_values(by='Season')

highest_diff_per_season
```

**Q16**: Vectorized Operation: Calculate the total number of games played by each team, whether they won or lost. Instead of iterating over the DataFrame, use the value_counts() function on the winning team and losing team columns separately, and then add the two Series together.

First, the number of wins and losses is counted using "value_counts()". Then, the results are combined using "add()", with missing values set to 0. The data is then converted into a DataFrame, and finally, it is sorted by the total number of games played.

```python
# Write your code here

# Calculate the number of games won by each team
winning_team_counts = df['Wteam'].value_counts()

# Calculate the number of games lost by each team
losing_team_counts = df['Lteam'].value_counts()

# Combine the two Series to get the total games played by each team
total_games_played = winning_team_counts.add(losing_team_counts, fill_value=0).astype(int)

# Convert to DataFrame for better presentation
total_games_played_df = total_games_played.reset_index()
total_games_played_df.columns = ['Team', 'TotalGames']

# Sort by total games played
total_games_played_df = total_games_played_df.sort_values(by='TotalGames', ascending=False)

total_games_played_df
```

Then, the data is grouped by season, and the "idxmax()" function is used to find the index of the game with the largest score difference in each season.
